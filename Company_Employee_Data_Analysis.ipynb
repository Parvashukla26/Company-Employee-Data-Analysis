```python
 department_salary = df.groupby('department')['salary'].sum()
 department_salary
```

```python
def number_format(department_salary):
    if department_salary >= 1e9:
        return f'{department_salary/1e9:.1f}B'
    elif department_salary >= 1e6:
        return f'{department_salary/1e6:.1f}M'
    elif department_salary >= 1e3:
        return f'{department_salary/1e3:.1f}K'
    else:
        return str(department_salary)
dept_salary = department_salary.apply(number_format)     
dept_salary
```

```python
department_salary.plot.bar(rot=45)
plt.title('Overall Salary by Department')
plt.xlabel('Department')
plt.ylabel('Total Salary')
plt.tight_layout() 
plt.show()
dept_salary
```
